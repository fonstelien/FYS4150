```python
def calc_energies(df):
    df['potential'] = -4*np.pi**2/np.sqrt(df['x']**2 + df['y']**2)
    df['kinetic'] = .5*(df['vx']**2 + df['vy']**2)
    df['total'] = df['potential'] + df['kinetic']
```

```python
usecols = ['1x','1y','1vx','1vy']
dfc1 = pd.read_csv('circular-n10.csv', usecols=usecols)
dfc2 = pd.read_csv('circular-n100.csv', usecols=usecols)
dfc3 = pd.read_csv('circular-n1000.csv', usecols=usecols)
dfc4 = pd.read_csv('circular-n10000.csv', usecols=usecols)
dfc1['reps'] = np.abs(np.sqrt(dfc1['1x']**2 + dfc1['1y']**2)-1)
dfc2['reps'] = np.abs(np.sqrt(dfc2['1x']**2 + dfc2['1y']**2)-1)
dfc3['reps'] = np.abs(np.sqrt(dfc3['1x']**2 + dfc3['1y']**2)-1)
dfc4['reps'] = np.abs(np.sqrt(dfc4['1x']**2 + dfc4['1y']**2)-1)

v = 2*np.pi
dfc1['veps'] = np.abs(np.sqrt(dfc1['1vx']**2 + dfc1['1vy']**2)-v)/v
dfc2['veps'] = np.abs(np.sqrt(dfc2['1vx']**2 + dfc2['1vy']**2)-v)/v
dfc3['veps'] = np.abs(np.sqrt(dfc3['1vx']**2 + dfc3['1vy']**2)-v)/v
dfc4['veps'] = np.abs(np.sqrt(dfc4['1vx']**2 + dfc4['1vy']**2)-v)/v
```

```python
usecols = ['1x','1vx']
dfd3 = pd.read_csv('drop-n1000.csv', usecols=usecols).dropna()
dfd4 = pd.read_csv('drop-n10000.csv', usecols=usecols).dropna()
dfd5 = pd.read_csv('drop-n100000.csv', usecols=usecols).dropna()

def v(r):
    return -2*np.pi*(2*(1/r-1))**.5
vr = v(dfd3['1x'][1:-1])
dfd3['eps'] = np.abs((vr-dfd3['1vx'][1:-1])/vr)
vr = v(dfd4['1x'][1:-1])
dfd4['eps'] = np.abs((vr-dfd4['1vx'][1:-1])/vr)
vr = v(dfd5['1x'][100:-1])
dfd5['eps'] = np.abs((vr-dfd5['1vx'][100:-1])/vr)
```

```python
fix, axes = plt.subplots(nrows=1, ncols=3, figsize=(21,6))

ax = axes[0]
ax.plot(dfc1.index[1:]/10*2, dfc1['reps'][1:], linewidth=THICK, label='$n=10^1$')
ax.plot(dfc2.index[1:]/100*2, dfc2['reps'][1:], linewidth=THICK, label='$n=10^2$')
ax.plot(dfc3.index[1:]/1000*2, dfc3['reps'][1:], linewidth=THICK, label='$n=10^3$')
ax.plot(dfc4.index[1:]/10000*2, dfc4['reps'][1:], linewidth=THICK, label='$n=10^4$')
ax.legend(loc='upper right')
ax.set_yscale('log')
ax.set_ylim(1E-9, 1E2)
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=.5))
ax.set_xlabel(r'Angle $\theta$ from the x-axis [rad.]', fontsize=MEDIUM)
ax.set_ylabel(r'Relative error of the $x_i,y_i$ position $\varepsilon = \frac{|r(\theta) - \sqrt{x_i^2 + y_i^2}|}{r(\theta)}$', fontsize=SMALL)

idx = dfc3['reps'][1:].argmax()
rmax = (dfc3['1x'].iloc[idx]**2 + dfc3['1y'].iloc[idx]**2)**.5
emax = -4*np.pi**2*(1/rmax-1)
xytext = (idx/500-.1,3E-4)
xy = (idx/500, dfc3['reps'].iloc[idx])
ax.annotate(f'$\Delta E_P/m  = $+{emax:.1e}', xytext=xytext, xy=xy, arrowprops={'arrowstyle':'->'})



ax = axes[1]
ax.plot(dfc1.index[1:]/10*2, dfc1['veps'][1:], linewidth=THICK, label='$n=10^1$')
ax.plot(dfc2.index[1:]/100*2, dfc2['veps'][1:], linewidth=THICK, label='$n=10^2$')
ax.plot(dfc3.index[1:]/1000*2, dfc3['veps'][1:], linewidth=THICK, label='$n=10^3$')
ax.plot(dfc4.index[1:]/10000*2, dfc4['veps'][1:], linewidth=THICK, label='$n=10^4$')
ax.legend(loc='upper right')
ax.set_yscale('log')
ax.set_ylim(1E-9, 1E2)
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=.5))
ax.set_xlabel(r'Angle $\theta$ from the x-axis [rad.]', fontsize=MEDIUM)
ax.set_ylabel(r'Relative error of the $v_x,v_y$ velocity $\varepsilon = \frac{|v(\theta) - \sqrt{v_{x,i}^2 + v_{y,i}^2}|}{v(\theta)}$', fontsize=SMALL)

idx = dfc3['veps'][1:].argmax()
vmax = (dfc3['1vx'].iloc[idx]**2 + dfc3['1vy'].iloc[idx]**2)**.5
emax = .5*vmax**2 - .5*4*np.pi**2
xytext = (idx/500-.1,3E-4)
xy = (idx/500, dfc3['veps'].iloc[idx])
ax.annotate(f'$\Delta E_K/m  = ${emax:.1e}', xytext=xytext, xy=xy, arrowprops={'arrowstyle':'->'})


ax = axes[2]
ax.plot(1-dfd3.index/1000/.178, dfd3['eps'], linewidth=THICK, label='$n=10^3$')
ax.plot(1-dfd4.index/10000/.1769, dfd4['eps'], linewidth=THICK, label='$n=10^4$')
ax.plot(1-dfd5.index/100000/.17672, dfd5['eps'], linewidth=THICK, label='$n=10^5$')
ax.legend(loc='upper right')
ax.set_yscale('log')
ax.set_xlabel('x', fontsize=MEDIUM)
ax.set_ylabel(r'Relative error of the velocity $\varepsilon = \frac{|v(x_i)-v_{x,i}|}{v(x_i)}$', fontsize=SMALL)


plt.show()
```

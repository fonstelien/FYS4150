```python
def calc_energies(df):
    df['potential'] = -4*np.pi**2/np.sqrt(df['x']**2 + df['y']**2)
    df['kinetic'] = .5*(df['vx']**2 + df['vy']**2)
    df['total'] = df['potential'] + df['kinetic']
```

```python
dfv100 = pd.read_csv('n100y100v.csv')
calc_energies(dfv100)
dfv1000 = pd.read_csv('n1000y100v.csv')
calc_energies(dfv1000)
dfe1000 = pd.read_csv('n1000y100e.csv')
calc_energies(dfe1000)
```

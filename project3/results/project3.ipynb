```python
def plot_trajectories(fname):
    df = pd.read_csv(fname)
    fix, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    axes[0].scatter(df['x'].iloc[0], df['y'].iloc[0], s=50)
    axes[0].scatter(df['x'].iloc[-1], df['y'].iloc[-1], s=50)
    axes[0].scatter(df['x'], df['y'], s=.1)
    axes[0].legend(['start', 'end'])

    axes[1].scatter(df['vx'].iloc[0], df['vy'].iloc[0], s=50)
    axes[1].scatter(df['vx'].iloc[-1], df['vy'].iloc[-1], s=50)
    axes[1].scatter(df['vx'], df['vy'], s=.1)
    axes[1].legend(['start', 'end'])
    plt.show()
    
```

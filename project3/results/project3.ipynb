```python
usecols = ['1x','1y','1vx','1vy']
dfc1 = pd.read_csv('circular-n10.csv', usecols=usecols)
dfc2 = pd.read_csv('circular-n100.csv', usecols=usecols)
dfc3 = pd.read_csv('circular-n1000.csv', usecols=usecols)
dfc4 = pd.read_csv('circular-n10000.csv', usecols=usecols)
dfc1['reps'] = np.abs(np.sqrt(dfc1['1x']**2 + dfc1['1y']**2)-1)
dfc2['reps'] = np.abs(np.sqrt(dfc2['1x']**2 + dfc2['1y']**2)-1)
dfc3['reps'] = np.abs(np.sqrt(dfc3['1x']**2 + dfc3['1y']**2)-1)
dfc4['reps'] = np.abs(np.sqrt(dfc4['1x']**2 + dfc4['1y']**2)-1)

v = 2*np.pi
dfc1['veps'] = np.abs(np.sqrt(dfc1['1vx']**2 + dfc1['1vy']**2)-v)/v
dfc2['veps'] = np.abs(np.sqrt(dfc2['1vx']**2 + dfc2['1vy']**2)-v)/v
dfc3['veps'] = np.abs(np.sqrt(dfc3['1vx']**2 + dfc3['1vy']**2)-v)/v
dfc4['veps'] = np.abs(np.sqrt(dfc4['1vx']**2 + dfc4['1vy']**2)-v)/v
```

```python
usecols = ['1x','1vx']
dfd3 = pd.read_csv('drop-n1000.csv', usecols=usecols).dropna()
dfd4 = pd.read_csv('drop-n10000.csv', usecols=usecols).dropna()
dfd5 = pd.read_csv('drop-n100000.csv', usecols=usecols).dropna()

def v(r):
    return -2*np.pi*(2*(1/r-1))**.5
vr = v(dfd3['1x'][1:-1])
dfd3['eps'] = np.abs((vr-dfd3['1vx'][1:-1])/vr)
vr = v(dfd4['1x'][1:-1])
dfd4['eps'] = np.abs((vr-dfd4['1vx'][1:-1])/vr)
vr = v(dfd5['1x'][100:-1])
dfd5['eps'] = np.abs((vr-dfd5['1vx'][100:-1])/vr)
```
